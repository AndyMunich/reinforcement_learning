### Definition of Reinforcement Learning
Reinforcement learning is the science of decision making with the goal to select actions which maximize the total future reward.
Given an observation (the current state of a game, the readings of a sensor, etc.) the policy maps the observation to an action.

RL is always about learning the policy, how this is done and what the policy actually is differs depending on the approach.
The policy can be a simple linear formula which multiplies each sensor signal with a weight, sums them up and the output is the motor torque.
It could also be a lookup table which says given the current observation/state take the following action.

### Reward, History, State and Markov State
Depending on the chosen action, a reward is given by the environment. The reward can be a +/- 1 at the end of a game while the rest of the time it is zero, or some feedback after each action.
In contrast to standard supervised learning where the supervisor already gives the expected output, in RL the reward might be a delayed signal which was a result of decisions done somewhere in the past. 
Sacrificing a queen early in a chess mach might be a bad move even if you dont loose immediately. 
Figuring out which decisions send you down the hill and assigning a negative reward to the corresponding decision is one of the big challenges in RL.

Non trivial problems often require the agent to not only look at the current observation but to have a memory/History. Saving the entire history often requires to much memory which is why some compressed form is saved, the so called state.
e.g. you can save the entire trajectory of a thrown object to calculate where it will move next or you simply save the current position and velocity.
A state which contains all relevant information is called **Markov** and does mean that the future is independent of the past given the current state.

### Observability
So far an observation and state are the same thing but this assumption only holds in environments with full observability. In reality the observation you make with your eyes might not be the environment state. Games like poker where you don't know the full environment are called partially observable but those will be covered later on. For now the agent state is just called state and contains all relevant information.

### Value function
The value function is a prediction of future reward and is used to evaluate states.

$v_\pi (s) = E_\pi [R_t + \gamma R_{t+1} + \gamma^2 R_{t+2} + ... | S_t = s]$ 

$\pi$: Policy <br>
$v_\pi$: value function for policy $\pi$ <br>
$R_t$: Reward at time step t <br>
$\gamma$: discount factor <br>

The discount factor is some number between 0 and 1 which discounts future rewards due to e.g. uncertainty in the environment model.

### Model
A model predicts what the environment will do next. It consists of two matrices. A **transition model** which describes with which probability the agent will land in which state given that it takes a specific action. <br>
$P^a_{s s'} = P[S_{t+1}=s' | S_t = s, A_t = a]$

The **reward model** estimates how much reward the agent gets in a specific state given a certain action. <br>
$R^a_s = E[R_{t+1} | S_t = s, A_t = a]$


### RL-Approaches
Value based algorithms only use the value function to evaluate the states and pick the action which results in a state with a higher value.

Policy based algorithms directly change the policy without evaluating the states.

Actor critics contain combine the two approaches above.

Model based agents use the environment model as an addition.
